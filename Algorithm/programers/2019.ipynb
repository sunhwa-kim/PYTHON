```python
from queue import PriorityQueue

def solution(food_times, k):
    if sum(food_times) <= k:
        return -1
    answer = 0
    q = PriorityQueue()
    for i in range(len(food_times)):
        q.put((food_times[i], i+1 ))
    sum_value = 0
    previous=0
    length = len(food_times)
    ####  k를 빼지 않고, sum_value 비교값으로 둔개 포인트
    while sum_value + ((q.queue[0][0] - previous)*length) <= k:  
            #  앞 접시까지 총 시간 + 다음 접시 다 먹는 시간 - 이전 걸린시간(비운 접시 동안) * 해당 접시 -1씩 감소하며 돈 횟수
        now = q.get()[0]  # sec
        sum_value += (now - previous)*length    # 가장 작은애 정해진 초당 접시를 비운경우를 제거 : 접시비운 것 = 회전수 =length
        length -= 1   # 최소값 빠지면 0 -> 시간 건너띄니까
        previous = now
        
    # 남은 음식 중 몇번 째 음식인지
    target = k - sum_value + 1
    length = len(q.queue)
    temp = (target - 1)//length
    result = sorted(q.queue, key=lambda x:x[1])  # 원래 index순서대로 바꾸고
    target -= temp * length
    return result[target - 1][1]

food_times = [3, 1, 2]
k = 5
solution(food_times, k)
```

```python
import re
from collections import Counter

def solution(str1, str2):
    text = re.compile('[^a-z]+')
    str1 = str1.lower()
    str2 = str2.lower()

    tmp1_lst = [ str1[i-2:i] for i in range(2,len(str1)+1) if not text.search(str1[i-2:i])]
    tmp2_lst = [ str2[i-2:i] for i in range(2,len(str2)+1) if not text.search(str2[i-2:i])]

    counter_1 = Counter(tmp1_lst)
    counter_2 = Counter(tmp2_lst)

    intersections = counter_1 & counter_2 # 교집합
    intersections = sum(list(intersections.values()))

    unions = counter_1 | counter_2 # 합집합
    unions = sum(list(unions.values()))
    if unions==0:
        answer = 65536
    else:
        answer = int(intersections/unions * 65536)

    return answer
solution(str1, str2)
```
