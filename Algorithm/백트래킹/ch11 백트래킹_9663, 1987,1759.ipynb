- N 의 범위가 적게 나온 것인지 체크 
    - 시간복잡도 dfs(재귀-stack) > bfs
- 완전탐색 수행 ->  DFS, BFS
    - 모든 경우의 수를 다 고려해야 하기 때문에 수가 꽤 커진다

```python
# 라이브러리 이용 예
from itertools import combinations

vowels = ('a', 'e', 'i', 'o', 'u')
l, c = map(int, input().split(' '))

# 가능한 암호를 사전식으로 출력해야 하므로 정렬 수행
array = input().split(' ')
array.sort()

# 길이가 l인 모든 암호 조합을 확인
for password in combinations(array, l):
    # 모음의 개수를 세기
    count = 0
    for i in password:
        if i in vowels:
            count += 1
    # 최소 한 개의 모음과 최소 두 개의 자음이 있는 경우 출력
    if count >= 1 and count <= l - 2:
        print(''.join(password))
```
