# LINKED LIST
- sperated datas connect, and manage = list() in python
- Node, pointer
- check : the last node has "node.next==None" in delete method 

```python
class Node:
    # data와 주소까지 인자로 넣어주는
    # next 들어가면 인자로 들어가고, 안 넣어줄 경우 None 처리 되게
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
        
class NodeCntr:
    
    def __init__(self, data):
        self.head = Node(data)
    
#     # for문 data 확인시 편리한....
#     def __iter__(self):
#     node = self.head
#     while node is not None:
#         yield node
#             node = node.next
    
    def ndshow(self):
        node = self.head
        while node:
            print(node.data)
            node = node.next
            
    def ndadd(self, data):
        if self.head is not None:
            node = self.head  #맨 앞 node를 알아야 다음걸 추가 하겠지
            while node.next:
                node = node.next
            node.next = Node(data)
        else:
            print('There is no data')
            
# 숫자일때만.. 순서...
    def ndinsert(self, data):
        idx = int(data)
        insrt_node = Node(data)

        node = self.head
        while node.data != int(data):
            node = node.next

        tmp = node.next 
        node.next = insrt_node
        insrt_node.next = tmp
    
    def nddel(self, data):
        if self.head =='':
            print('첫 번째 node가 없습니다.')
            return
        
        # header일 때 while문 True로 돌리다 보니 node에 next가 없어 NoneType 에러 발생
        if self.head.data==data:
            tmp = self.head
            self.head = self.head.next
            del(tmp)
        else:
            idx = True
            node = self.head
            while idx:
                if node.next.data == data:
                    idx=False
                else:
                    node = node.next
            tmp = node.next
            node.next = node.next.next  # 마지막은 next=None일 테니까 알아서 들어가겠지
            del(tmp)
    
    def ndsearch(self, data):
        if self.head is None:
            print('There is no data')
            return
        else:
            node = self.head
            while node is not None:
                if node.data == data:
                    print(node.data)
                    return
                else:
                    node = node.next
            print('No search data')
```

## DOUBLE LINKED LIST
- 양방향 연결로 NODE 탐색시 양쪽 모두 가능

```python
class DoubleNode:
    def __init__(self, data,prev=None, next=None):
        self.prev = prev
        self.data = data
        self.next = next
```

```python
class DoubleNodeCntr:
    
    def __init__(self, data):
        self.head = DoubleNode(data)
        self.tail = self.head  #뒤에서부터 찾으려면 head만 바뀌면 돼
        
    def Dbl_ndshow(self):
        node = self.head
        while node.next:
            print(node.data)
            node = node.next
        print(node.data)
        
    def Dbl_ndReverse(self):
        node = self.tail
        while node.prev:
            print(node.data)
            node = node.prev
        print(node.data)
        
    def Dbl_ndadd(self, data):
        node = self.head
        # 일단 뒤로 찾아가 -> next,prev 저장 -> tail변경
        while node.next:
            node = node.next
        new_node = DoubleNode(data)
        new_node.prev = node
        node.next = new_node
        self.tail = new_node
    
    def Dbl_nddel(self,data):
        if self.head.data == data:
            tmp = self.head
            self.head = self.head.next
            self.head.prev = None
            del(tmp)
            return
        else:
            node = self.head.next
            while node:
                if node.data == data:
                    if node == self.tail:
                        self.tail = node.prev
                        node.prev.next = node.next
                        del(node)
                        return
                    else:
                        node.prev.next = node.next
                        node.next.prev = node.prev
                        return
                else:
                    node = node.next
                    
    def Dbl_search(self, data):
        midnd = int(self.tail.data/2) - data
        if midnd >=0 :
            node = self.head
            while node:
                if node.data == data:
                    #print('head')
                    #print(node.data)
                    #break
                    return node.data
                else:
                    node = node.next
        else:
            node = self.tail
            while node:
                if node.data == data:
                    #print('tail')
                    #print(node.data)
                    #break
                    return node.data
                else:
                    node = node.prev
```
